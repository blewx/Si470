```python
import pandas as pd

# Point this to your CSV
file_path = "all_countries.csv"

# Load
df = pd.read_csv(file_path)

# Keep the parent category in the column names
df["Category_Subcategory"] = (
    df["Category"].astype(str).str.strip()
    + ": "
    + df["Subcategory"].astype(str).str.strip()
)

# Pivot: countries as rows, subcategories as columns
pivot = df.pivot_table(
    index="countryISO3",
    columns="Category_Subcategory",
    values="hoursPerDayCombined",  # main metric to compare
    aggfunc="mean"                 # or "sum" if that's more appropriate
)

# Make the header look clean (remove that extra top label)
pivot.columns.name = None

# Optional: Ssort columns alphabetically
pivot = pivot.reindex(sorted(pivot.columns), axis=1)

# If you prefer countryISO3 as a normal column:
pivot_out = pivot.reset_index()

# Save or inspect
pivot_out.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import numpy as np

# --- Convert to numeric matrix; handle missing values ---
X = pivot.fillna(0.0).to_numpy(dtype=float)   # shape: (n_countries, n_features)

print(X)
```

```python
# --- Fit KMeans ---
n_clusters = 6
kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init="auto").fit(X)

# --- Get cluster labels ---
labels = kmeans.labels_
print(kmeans.labels_)
```

# Append cluster # to data frame

```python
# Add cluster labels to pivot_out
pivot_out["cluster"] = labels

# Print as a plain table
pivot_out.head()  # show first 20 rows, remove limit if you want full table

```

```python
###### Ensure cluster column exists
pivot_out["cluster"] = labels

# Count countries per cluster
cluster_sizes = pivot_out.groupby("cluster").size().rename("n_countries")

# Average values per cluster
cluster_summary = pivot_out.groupby("cluster").mean(numeric_only=True)

# Add the counts column
cluster_summary = cluster_summary.merge(cluster_sizes, left_index=True, right_index=True)

# Add a list of countries for each cluster
countries_in_cluster = pivot_out.groupby("cluster")["countryISO3"].apply(list).rename("countries")

# Merge into summary
cluster_summary = cluster_summary.merge(countries_in_cluster, left_index=True, right_index=True)

# Reset index so cluster is a column
cluster_summary = cluster_summary.reset_index()

# Print nicely in table format
cluster_summary.head(n_clusters)

```

```python
from sklearn.metrics import silhouette_score

for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    print(f"k={k}, silhouette score={score:.3f}")
```

```python
from sklearn.metrics import silhouette_score

for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    print(f"k={k}, silhouette score={score:.3f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertias = []
K_range = range(1, 15)

for k in K_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0).fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(K_range, inertias, 'bo-')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (within-cluster sum of squares)")
plt.title("Elbow Method")
plt.show()
```

```python
import wbdata
import pandas as pd

# --- Indicator: GNI per capita (current US$) ---
indicator = {"NY.GNP.PCAP.CD": "GNI_per_capita"}

# --- Fetch full dataset ---
df = wbdata.get_dataframe(indicator).reset_index()

# --- Keep only most recent year ---
latest_year = df["date"].max()
df_latest = df[df["date"] == latest_year].drop(columns=["date"])

# --- Load your ISO mapping (from all.csv) ---
iso_map = pd.read_csv("../all.csv")[["name", "alpha-3"]]

# --- Merge to keep only valid countries ---
df_latest = df_latest.merge(iso_map, left_on="country", right_on="name", how="inner")

# --- Keep only ISO3 + GNI ---
df_latest = df_latest[["alpha-3", "GNI_per_capita"]].rename(columns={"alpha-3": "ISO3"})

# --- Save + preview ---
df_latest.to_csv("world_bank_gni_countries.csv", index=False)
print(f"Latest year in dataset: {latest_year}")
print(df_latest.head(30))
```

```python
import wbdata
import pandas as pd

# --- 1) Fetch World Bank indicator (GNI per capita, current US$) ---
indicator = {"NY.GNP.PCAP.CD": "GNI_per_capita"}
df = wbdata.get_dataframe(indicator).reset_index()  # columns: country, date, GNI_per_capita

# --- 2) For each country, keep the most recent NON-NULL value (backtrack years if needed) ---
df_nonnull = (
    df.dropna(subset=["GNI_per_capita"])
      .sort_values(["country", "date"])            # ascending by date
)
latest_per_country = df_nonnull.groupby("country", as_index=False).tail(1)  # last = most recent non-null

# --- 3) Load your ISO mapping (real countries) and merge ---
# all.csv should have at least columns: name, alpha-3
iso_map = pd.read_csv("../all.csv")[["name", "alpha-3"]]

out = (
    latest_per_country
    .merge(iso_map, left_on="country", right_on="name", how="inner")  # keeps only real countries
    .rename(columns={"alpha-3": "ISO3"})
    .loc[:, ["ISO3", "GNI_per_capita"]]
    .sort_values("ISO3")
    .reset_index(drop=True)
)

# --- 4) Save / preview ---
out.to_csv("world_bank_gni_latest_nonnull_iso3.csv", index=False)
print(out.head(20))
```
